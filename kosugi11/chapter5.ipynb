## 5.1 確率過程の定義

用語
* **確率過程**とは、確率変数であり、時刻 $t$ に依存するものだ。
  * 記号 $X(t)$, $\lbrace X(t) \rbrace$ などで表す。
  * どうして stochastic process が「確率」過程と呼ばれるのか気になる。
  * 確率空間 $(\Omega, \mathscr{F}, P)$ と時刻の集合 $T \subset \mathbb{R}$ があって、
    $t \in T$ に対してこの確率空間上の確率変数 $X_t = X(t)$ が対応しているといっている。

* **標本関数**とは、確率過程の取りうる値だ。
  * 記号 $x(t)$ などで表す。
  * 確率過程 $\lbrace X(t) \rbrace$ で $\omega \in \Omega$ を固定して得られる $t$ を変数とする関数
    $t \mapsto X_t(\omega)$ として見ている。

 

## 5.2 ランダムウォーク

ランダムウォークが関係する現象は、興味のある期待値が直感に反することが多いようだ。

* ここで述べるのは一次元単純**ランダムウォーク**。
* 時刻ゼロで原点に位置するとする。
* 移動する時刻は $t$ が整数であるときとし、移動時間ゼロで位置を更新するものとする。
* 移動時刻において、確率 $p\ (0 < p < 1)$ で $1$ だけ、確率 $q = 1 - p$ で $-1$ だけ移動するものとする。
  * $p = q = \dfrac{1}{2}$ のときのランダムウォークを対称ランダムウォークという。以下、これを取り扱う。


* 確率変数 $X_i$: 時刻 $i \in \mathbb{Z}$ における移動者の変位。
* 部分和 $S_n = X_1 + \dotsb + X_n$ がランダムウォークを表すと解釈する。
  * $S_0 = 0$ と約束する。
  * $S_n$ の取りうる値は $n, n - 2, n - 4, \dotsc, -(n - 4), -(n - 2), -n$ のように対称になる。
    * 移動者が奇数時刻に原点に位置することはない。

時刻 $n$ に点 $S_n = n - 2k$ に移動者がいる確率は次で表される：
$$
P(S_n) = \frac{1}{2^n}
\begin{pmatrix}n \\ k\end{pmatrix}.
$$

* **ランダムウォークの道**とは、$i = 0, 1, \dotsc$ について
  $S_i$ のとる値が $s_i$ であるときのグラフ $(i, s_i)$ である。
  ただし隣接する移動時刻間の各点を線分で結ぶものとする。
  

* 補題 5.1. **鏡像原理**

  仮定：点 $A(a, s_a)$ と 点 $B(b, s_b)$ について $a < b$, $s_a s_b > 0$ とする。
  
  結論：A から B へ至る道のうち、$\exists n (s_n = 0)$ なものの総数は、
  点 $A'(a, -s_a)$ と点 B へ至る道の総数と等しい。


* 命題 5.2. 道の総数計算式

  * $m > 0$ という条件をつける。
  * $N_{n,m}$: 原点から点 $(n, m)$ へ至る道の総数とする。組み合わせの計算により求められる。
  * $\forall i \in \lbrace 1, \dotsc, n - 1\rbrace (s_i > 0)$ となる道の総数は、補題 5.1. を応用して簡単に求められる。
  


### 5.2.1 ランダムウォークの道

* ランダムウォークの道が原点に戻る場合の確率に興味がある。
* 確率がゼロでないのは時刻が偶数である必要があるので、時刻を $2n$ とおくことにする。


* 命題 5.3. 特殊なランダムウォーク道の総数

  1. $(\forall i \in \lbrace 1, \dotsc, 2n - 1\rbrace (s_i > 0)) \land (s_{2_n} = 0)$ となる道の総数
  2. $(\forall i \in \lbrace 1, \dotsc, 2n - 1\rbrace (s_i \ge 0)) \land (s_{2_n} = 0)$ となる道の総数
  
  
あとの記述を助けるために数列を定義しておく。
$$
\begin{align*}
u_0 &= 1, \quad u_{2n} = \dfrac{1}{2^{2n}}\begin{pmatrix}2n \\ n \end{pmatrix},\\
f_0 &= 0, \quad f_{2n} = \dfrac{1}{2n} u_{2n - 2},
\end{align*}
$$


* 命題 5.4. 数列 $f_n$ の一般項

  $f_n$ は $u_{2n}$ の階差であるようだ。


* 定理 5.5. いろいろな道の確率を数列 $u_{2n}, f_{2n}$ で表す

  * $u_{2n}$: 時刻 $2n$ に原点にいる確率＝時刻 $2n$ まで最初しか原点にいない確率＝時刻 $2n$ までずっと $s_i > 0$ である確率
  * $f_{2n}$: 時刻 $2n$ に出発以来初めて原点にいる確率＝時刻 $2n - 1$ で初めて $s_{2n - 1} = -1$ にいる確率
  
  証明：命題 5.3. をうまく変形することでどの場合も別個に示せる。
  

* 系 1. 時刻 $\forall r \le n$ で出発以来初めて原点にいる確率 $f_{2r}$ および
  最終時刻 $2n$ に原点にいる確率を和で考えることで、次の関係が得られる：
  $$
  u_{2n} = \sum_{r = 1}^{n} f_{2r} u_{2n - 2r}.
  $$

### 5.2.2 滞在時間と逆正弦法則

* 定理 5.6. 移動者の滞在時間の確率
  * 時間間隔を $0$ から $2n$ とし、
  * $p_{2k, 2n}$: 移動者が正側 $(s_k > 0)$ に $2k$ 時間（負側に $(2n - 2k)$ 時間）いる確率
  とする。
  
  $$
  p_{2k, 2n} = u_{2k} u_{2n - 2k}
  $$
  
  証明
  
  1. $k = n$ のとき
  
     * 移動者がずっと正側にいる確率 $p_{2n, 2n}$ は定理 5.5. で求めたように $u_{2n}$ に等しい。
     * $u_{2n} = u_{2n} \cdot 1 = u_{2n} u_0 = p_{2n, 2n}$ であるので、成り立っている。
     
  2. $k = 0$ のとき
  
     * 移動者がずっと負側にいる確率 $p_{0, 2n}$ は対称性により 1. と同じ $u_{2n}$ である。
     * $u_{2n} = 1 \cdot u_{2n} = u_0 u_{2n} = p_{0, 2n}$ であるので、成り立っている。
  
  3. $k = 1, \dotsc, n - 1$ のとき
  
     * 出発してから初めて x 軸に戻る時刻を $2r\ (r \ne 0)$ とする。
     * 正側から戻る場合と負側から戻る場合に分けて考える。
     
     * 正の場合
       * 正側滞在時間帯は滞在時間が $2r$ の部分と $2k - 2r\ (\ge 0)$ の部分とに別れる。
       * 前者の確率は定理 5.5. より $f_{2r}$ の半分の $\dfrac{1}{2}f_{2r}.$
       * 総合して、正の場合の確率は次になる：
         $$
         \frac{1}{2}f_{2r} p_{2k - 2r, 2n - 2r}.
         $$
         
     * 負の場合
       * 正側滞在時間帯はすべて $2r \ge 0$ 以降に含まれ、$2k \le 2n - 2r$ が成り立つ。
       * 正の場合と同様に考えて、この場合の確率は次になる：
         $$
         \frac{1}{2} f_{2r} p_{2k, 2n - 2r}
         $$
         
     以上より、求める確率は次に等しい：
     $$
     p_{2k,2n} = \frac{1}{2} \sum_{r = 1}^k f_{2r} p_{2k - 2r, 2n - 2r} + 
     frac{1}{2} \sum_{r = 1}^{n - k} f_{2r} p_{2k, 2n - 2r}.
     $$
     
  以降は $n$ に関する数学的帰納法によって $p_{2k, 2n} = u_{2k} u_{2n - 2k}$ を示す。
  * 級数を $u_{K}$ の形に表すのに系 1. を用いる。


* 定理 5.7. **逆正弦法則**

  * $X$: ランダムウォークの正側滞在時間の割合を表す確率変数
  
  この密度関数 $f(x)$ は次で表される：
  $$
  f(x) = 
  \begin{cases}
  \dfrac{1}{\pi \sqrt{x (1 - x)}},&\quad 0 < x < 1\\
  0,&\quad\text{ otherwise}
  \end{cases}
  $$
  
  時刻 $0< \alpha < \beta < 1$ に対して、正側滞在時間の割合がここにいる確率は次で表される：
  $$
  P(\alpha < X < \beta) = \frac{2}{\pi}\left(\arcsin\sqrt{\beta} - \arcsin\sqrt{\alpha}\right).
  $$
  
  証明：
  1. $p_{2k,2n}$ を Stirling の公式で評価する。
  2. それを用いて正側滞在時間の割合 $k/n$ における範囲 $\alpha < k/n < \beta$ にわたる和を評価する。
  3. 割合 $k/n$ を $x$ と捉えて、和を積分に置き換えて考える。
  4. この積分が結論の密度関数と確率の値の両方を導くことを示す。


```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
x = np.arange(0.01, 1, 0.01)
plt.xlim(0, 1)
plt.plot(x, 1/(np.pi * np.sqrt(x * (1 - x))),
         1/2, 2/np.pi, 'o')
```

### 5.2.3 破産の問題

仮定：
* $a, z \in \mathbb{N}$
* プレイヤー A と B の所持金はそれぞれ $z,\ a - z$ ドル。
* ゲームをするたびに敗者が勝者へ 1 ドル支払う。
* 各ゲームで A が勝つ確率を $p$ とする。
* 便宜上いつものように $q = 1 - p$ とおく。

これはランダムウォークの応用になる。
* 出発点が $x = z$ で
* $x = 0$ と $x = a$ に「壁」がある

ようなランダムウォーク。

#### A が破産する確率
記号を導入する。
* $q(z)$: A が破産する確率
* $p(z)$: A が完全勝利する確率

これらの確率はゲーム開始時の所持金に依存する。
以下、特に断らなければ A 視点での勝ち負けについて述べているものとする。

* $q(z) = p q(z + 1) + q q(z - 1)$ についてはもう少し説明してほしい。
 * 「所持金を 1 ドルだけ多く持って連戦した場合の破産確率」と
 * 「所持金を 1 ドルだけ少なく持って連戦した場合の破産確率」との関係か。
* $q(0) = 1$: A が鉄砲ならばすでに破産状態なので。
* $q(a) = 0$: これは B が鉄砲の場合。

上の差分方程式をなんとか解く。これは知識がないのでなんとかしたい。
要点をまとめると：
* $p \ne q$ と $p = q$ とに場合分けして解く。
* どちらも $q(z)$ の一般解が求まれば、$p(z)$ のそれは式中の $p, q, z$ を $q, p, a - z$ にそれぞれ置き換えたものだ。
* どちらの場合も $p(z) + q(z) = 1$ が成り立つ。


賭金を変えることの考察：
* 賭金を $k\ (k \ne 0)$ 倍することは所持金を $1/k$ 倍することに相当する。
  $p(z), q(z)$ の式中の $a, z$ をそれぞれ $a/k, z/k$ に置き換えたものがレート変化後の確率を表す。
* 変化前と変化後の $q(z)$ の比をみれば、破産確率の変動がどちらなのかがわかる。
* 結論だけいうと、レートを上げるのは A の勝利確率 $p$ が低いときに A にとって有利に働く。

#### ゲームの期待継続時間
$D_z$ を期待継続時間（一方が破産するまでの総ゲーム回数）とする。

さきほどと同様に考えて、次の差分方程式と境界条件がある：
$$
D_z = pD_{z + 1} + qD_{z - 1} + 1,\quad D_0 = 0,\quad D_a = 0.
$$

要点をまとめると：
* $p \ne q$ と $p = q$ とに場合分けして解く。
* 特に $p = q$ のときの継続時間が $z(a - z)$ なので、
  コイントスのような公平なゲームを A の所持金が 1 ドルであるとしても、
  ほとんど B の所持金と同じ値の回数が期待継続時間になってしまう。

## 5.3 Brown 運動

ランダムウォークの連続的なものを考えていく。時間と空間のスケールを無限に細かくする。

単純対称ランダムウォークの期待値と分散を書いていなかった。
$$
E[X_i] = 0,\quad V(X_i) = 1.
$$
$$
E[S_n] = 0,\quad V(S_n) = n.
$$
ランダムウォークの道の極限は、時間スケールと空間スケールをそれぞれ $1/N$, $1/\sqrt{N}$ にとると、
de Moivre-Laplace の定理により、正規分布に収束する。

* 定理 5.8. 

  $$
  Y_t^{(N)} = \frac{1}{\sqrt{N}}S_{\lfloor Nt \rfloor}
  $$
  
  このスケーリングによって得られる確率変数の確率の極限は正規分布 $N(0, t)$ に従う：

  $$
  \lim_{N \to \infty} P\left(a \le Y_t^{(N)} \le b\right)
  = \frac{1}{\sqrt{2\pi t}}\int_a^b\!
      \exp\left(-\frac{x^2}{2t}\right)
      \,\mathrm{d}x.
  $$
  
  証明
  
  1. 中心極限定理によると $\dfrac{S_n}{\sqrt{n}}$ は $N(0, 1)$ に収束する。
  2. 
     $$Y_t^{(N)} = \dfrac{S_{\lfloor Nt \rfloor}}{\sqrt{\lfloor Nt \rfloor}} 
     = \sqrt{t}\dfrac{S_{\lfloor Nt \rfloor}}{\sqrt{\lfloor Nt \rfloor}}.
     $$
  3. $n = \lfloor Nt \rfloor$ を 1. に代入して
     $\dfrac{S_{\lfloor Nt \rfloor}}{\sqrt{\lfloor Nt \rfloor}}$ は $N(0, 1)$ に収束する。
  4. $Nt - 1 \lt n \le Nt$ より、次の挟み撃ちが成り立つ：
     $$
     \frac{1}{Nt} \le \frac{1}{\lfloor Nt \rfloor} \lt \frac{1}{Nt - 1}.
     $$
  5. 以上より 2. の右辺は $N(0, t)$ に収束する。


$Y_t$: $Y_t^{(N)}$ の極限としての確率変数とする。
この（連続的）確率過程を 1 次元 Brown 運動という。

* 定義 5.1. **Brown 運動**

  確率過程 $W(t),\: t \le 0$ が Brown 運動であるとは、この確率過程が次の条件をみたすことをいう：
  
  * $W(0) = 0.$
  * $\forall n \in \mathbb{N},$
    $t_1 < t_2 < \dotsb < t_n$ であるような $\forall t_i \in [0, \infty)$ について
    次の各確率変数が独立である：
    
    $$
    W(t_1),\ W(t_2) - W(t_1),\ \dotsc, W(t_n) - W(t_{n - 1})
    $$
    
  * $\forall s, t \in (0, \infty)$ に対して $W(t) - W(s)$ の分布が $N(0, \left|t - s\right|)$ に従う


* 命題 5.9. Brown 運動の増分（差分）の自乗の期待値

  $$
  \forall s, t \in (0, \infty) \left(E\left[(W(t) - W(s))^2\right] = t - s.\right)
  $$


* 定理 5.10. Brown 運動の道 $\{W(t)\}$ に関する性質

  運動の時間間隔を $T$ とする。このとき、

  * ほとんど確実に (a.s.) 連続である。 
  * すべての $t \in T$ で微分不可能である。
  * 時間間隔の二次変動の期待値がほとんど確実に $T$ になる：
  
    * $0 = t_0 < \dotsb < t_n = T$: 区間の細分
    * $\delta = \displaystyle \max_k\{t_k - t_{k - 1}\}$: 細分幅の最大値
    * $E_\delta$: この細分での期待値
    $$
    \lim_{\delta \to 0}
      E_\delta \left[
        \sum_{k = 1}^n
          \left(W(t_k) - W(t_{k - 1})\right)^2
        \right]
      = T.
    $$
    
  証明がないので、どこかから持ってくるか？


## 5.4 Markov 連鎖

* 定義 5.2. **Markov 連鎖**、**推移確率**

  * Markov 連鎖とは、
  
    * 確率変数列 $X_n\ (n = 0, 1, 2, \cdots)$ であり、
    * 各 $X_n$ が高々可算集合 $S$ の中に値を取り、
    * 条件付き確率分布 $P(X_{n + 1} = j \mid X_0 = i_0, \cdots X_n = i_n)\quad(i_0, \cdots, i_n \in S)$ が、
    * 条件付き確率分布 $P(X_{n + 1} = j\mid X_n = i_n) \quad(j \in S)$ と一致するものをいう。

  * 推移確率とは、次で定義される確率関数である：
    $$
    p_{ij} = P(X_{n + 1} = j\mid X_n = i).
    $$
    
    
* 職業病のようなもので状態推移ではなく状態遷移といってしまうが、英語ならどちらも transition だから問題ない。
  岩波数学辞典も「推移」派。


* 一次元対称単純ランダムウォーク $(S_n, \mathbb{N}, p_{ij})$ は Markov 連鎖であり、各 $p_{ij}$ は $0$ か $1/2$ しかとらない。

以降、有限 Markov 連鎖 $(X_n, \{1, \dotsc, N\}, p_{ij})$ を考える。

用語
* **推移行列**とは行列であり、Markov 連鎖の確率関数 $p_{ij}$ を $(i, j)$ 成分とするものである。
  * 各行の成分の和が 1 になる。なぜなら各行は、その表す状態 $i$ からすべての状態に推移する確率の値を並べたものだからだ。
  
記号
* $p_{ij}^{(n)}$: $n$ 回の推移で状態 $i$ から $j$ へ移る確率とする。
  * $p_{ij}^{(1)} = p_{ij}.$
  * $p_{ij}^{(2)} = \displaystyle \sum_{r = 1}^N p_{ir} p_{rj}.$

* 定理 5.11. **Chapman-Kolmogorov の定理**

  $$
  p_{ij}^{(n)} = \sum_{r = 1}^N p_{ir}^{(m)} p_{rj}^{(n - m)}, 1 \le m \le n - 1.
  $$
  
  特に $m = 1$ とすると：
  $$
  p_{ij}^{(n)} = \sum_{r = 1}^N p_{ir} p_{rj}^{(n - 1)}.
  $$

記号
* $Q^{(n)}$: 正方行列であり $(i, j)$ 成分が $p_{ij}^{(n)}$ であるものとする。

  * $Q^{(n)} = Q Q^{(n - 1)} = \cdots = Q^n.$

Markov 連鎖の確率分布は、初期分布 $X_0$ と推移行列 $Q^{(n)}$ により決まる。


* 定義 5.3. **吸収状態**

  * Markov 連鎖の状態 $i \in S$ が吸収状態であるとは、$p_{ii} = 1$ であるものをいう。
  * いったんこの状態に陥ると、他の状態に推移することはない。

本書では例として「酒に酔った男の吸収状態あり吸収壁ありランダムウォーク」を挙げている。

Markov 過程の興味は推移確率 $p_{ij}^{(n)}$ の極限らしい。

* 定義 5.4. **再帰的** (1)

  状態 $i \in S$ が

  * 再帰的であるとは、状態 $i$ から出発して、少なくとももう一度状態 $i$ になる確率が 1 であることをいう。
    * そうでないものを**非再帰的**であるという。
  * **正再帰的**であるとは、状態 $i$ に戻るまでの平均時間が有限であることをいう。


* 定義 5.5. **再帰的** (2)

  Markov 連鎖 $X_n$ が
  
  * 再帰的であるとは、$\forall i \in S (\text{i is recurrent})$ であることをいう。
  * **正再帰的**であるとは、$\forall i \in S (\text{i is positive recurrent})$ であることをいう。


* 定義 5.6. **推移可能**、**既約**

  * $i, j \in S$ について $i$ が $j$ に推移可能であるとは、$\exists m (p_{ij}^{(m)} > 0)$ であることをいい、
    記号 $i \to j$ で示す。
    
  * Markov 連鎖 $(X_n, S, p_{ij})$ が既約であるとは、
    $\forall i, j \in S ((i \to j) \land (j \to i))$ であることをいう。

Markov 連鎖が既約であれば、

1. すべての点が再帰的であるか、
2. すべての点が非再帰的であるか

のどちらか一方がなりたつらしい。


* 定義 5.7. 再帰的状態の**周期**

  状態 $i \in S$ を再帰的状態とすると $p_{ii}^{(n)} > 0$ なる $n \le 1$ がある。
  このような $n$ の最大公約数をこの状態の周期といい、記号 $d(i)$ で示す。
  
  * 特に $d(i) = 1$ のとき、この状態は**非周期的**であるという。
  * それ以外のときには**非周期的**であるという。


* 定義 5.8. 既約な Markov 連鎖の**周期**

  既約 Markov 連鎖はすべての状態が同じ周期 $d$ であるので、
  これを既約 Markov 連鎖の周期という。


* 定理 5.12. **Markov 連鎖の極限定理**

  ノート：既約かつ正再帰的かつ非周期的 Markov 連鎖 $(X_n, S = (1, \dotsc, N), p_{ij})$ の
  $n$ 段推移行列 $Q^{(n)}$ の極限に関する記述。
  
  **定常分布**という次の性質を有する実数の組 $(\pi_1, \dotsc, \pi_N)$ が存在して、
  上述の行列極限はこの組を $N$ 個縦に並べたものに収束するというのが本定理だ：
  
  $$
  \begin{align*}
  \pi_j &= \sum_{i = 1}^N \pi_i p_{ij},\\
  \sum_{j = 1}^N \pi_j &= 1.
  \end{align*}
  $$


以下、Markov 連鎖の例。

### 例 5.1. コンビニ在庫量

1. あるコンビニのある商品の在庫と需要の確率分布が与えられている。
2. 補充方式から、この商品の在庫量は Markov 連鎖に従う。
3. 状態遷移図を描き、推移行列 $Q$ を得る。
4. 極限定理が適用できる Markov 連鎖なので、定常分布 $\boldsymbol{\pi} = (\pi_1, \pi_2, \pi_3)$ を求めることにする。
5. 方程式 $\boldsymbol{\pi} Q = \boldsymbol{\pi}$ を解く。このとき解が一意的であるには $\sum \pi_j = 1$ が必要。
6. 定常分布が得られると需要発生直後の在庫の分布が得られる。

### 例 5.2 吸収状態のある Markov 分布の例

商品が吸収状態のある Markov 連鎖をもつ場合。

|状態|意味|→新酒|→古酒|→販売|→廃棄|
|---:|----|---:|---:|---:|---:|
|1   |新酒|0.0 |0.6 |0.3 | 0.1|
|2   |古酒|0.0 |0.2 |0.5 | 0.3|
|3   |販売|0.0 |0.0 |1.0 | 0.0|
|4   |廃棄|0.0 |0.0 |0.0 | 1.0|

この推移行列 $Q$ を左上 2x2 行列 $Q_{11}$ と右上 2x2 行列 $Q_{12} に分割して吸収状態に至る確率を求める。

1. $Q = \begin{pmatrix} Q_{11} & Q_{12}\\ O & E \end{pmatrix}$
   だから
   $Q^n \to \begin{pmatrix} Q_{11}^n & R\\ O & E \end{pmatrix}\ (n \to \infty)$
   の形になる。
   
   $R$ が吸収状態に至る確率をもつ行列である。
   
2. 一方 $Q_{11}^n \to O\ (n \to \infty)$ なので次がいえる：
   $$
   \begin{pmatrix} Q_{11} & Q_{12}\\ O & E \end{pmatrix}
   \begin{pmatrix} O & R\\ O & E \end{pmatrix}
   =
   \begin{pmatrix} O & R\\ O & E \end{pmatrix}
   $$
3. つまり $Q_{11} R + Q_{12} = R$ が成り立つ。二次行列の演算なので $R$ は容易に求まる。 


```python
import numpy as np

Q_11 = np.array([[0.0, 0.6],
                 [0.0, 0.2]])
Q_12 = np.array([[0.3, 0.1],
                 [0.5, 0.3]])
E = np.eye(2)
R = np.linalg.inv(E - Q_11) @ Q_12; R
```
